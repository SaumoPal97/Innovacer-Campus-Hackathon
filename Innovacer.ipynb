```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dfs = pd.ExcelFile('IIT Kharagpur _ Campus Hiring - Problem Statement.xlsx')
```

```python
print dfs.sheet_names
df1 = dfs.parse('Data')
df1 = df1.drop([0,1])
df1 = df1.reset_index(drop=True)
df_names = df1['aco_name']
df1.drop(['aco_name','Unnamed: 5','py'], inplace=True, axis=1)
df1
```

```python
df1 = df1[df1['per_capita_exp_total_py'].notnull()]
sns.distplot(df1['per_capita_exp_total_py'])
```

```python
from scipy.stats import boxcox
def normalizing(X):
    df1[X][df1[X] > df1[X].quantile(0.99)] = df1[X].quantile(0.99)    
#   df1[X], _ = boxcox(df1[X])
features = ['per_capita_exp_total_py']
for i in features:
    normalizing(i)
sns.distplot(df1['per_capita_exp_total_py'])
```

```python
print df1.shape
print df1.isnull().sum(axis=0)
```

```python
df1.drop(['aco44','aco43','aco12'], inplace=True, axis=1)
print df1.shape
df1
```

```python
df1['qualscore'].plot.hist()
df1['qualscore'].fillna(df1['qualscore'].mode()[0], inplace=True)
```

```python
df1 = df1.dropna(subset=['n_ben_race_asian','n_ben_race_hisp','n_ben_race_native'])
print sorted(df1.isnull().sum(axis=0))
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
aco_num = df1['aco_num']
target = df1['per_capita_exp_total_py']
aco_state = df1['aco_state']
df1.drop(['aco_num','per_capita_exp_total_py','aco_state'], inplace=True, axis=1)
```

```python
from sklearn.metrics import r2_score
def multAcc(pred, dtrain):
    label = dtrain.get_label()
    acc = r2_score(label, pred)
    return 'maccuracy', acc
```

```python
final_features = [x for x in df1.columns if x not in ['met_qps','n_eta']]
print final_features
df1 = df1.replace('-',np.nan)
df1 = df1.astype('float')
df0 = df1
X_train, X_valid, y_train, y_valid = train_test_split(df1, target, train_size = 0.8, random_state = 2018)
dtrain = xgb.DMatrix(data=X_train[final_features], label=y_train)
dvalid = xgb.DMatrix(data=X_valid[final_features], label=y_valid)
watchlist = [(dtrain, 'train'),(dvalid, 'eval')]
```

```python
params ={'eta':0.1,'seed':0,'subsample':1,'colsample_bytree':0.8,'objective':'reg:linear','max_depth':8,'min_child_weight':1}
r1 = xgb.train(params, dtrain, 1000, watchlist, maximize=True, verbose_eval=20, early_stopping_rounds=100, feval=multAcc)
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20,15
xgb.plot_importance(r1)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
y_pred = regr.predict(X_valid)
print("Mean squared error: %.2f"
      % mean_squared_error(y_valid, y_pred))
for k in (list(zip(regr.coef_, X_train.columns))):
    print k
```
